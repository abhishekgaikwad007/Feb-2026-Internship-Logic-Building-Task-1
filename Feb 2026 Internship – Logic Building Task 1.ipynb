```python
marks = [45, 78, 90, 33, 60]
pass_count = 0
fail_count = 0
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1
print("Total Passed Students:", pass_count)
print("Total Failed Students:", fail_count)
```

```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = []
for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)
print("Cleaned Names List:", cleaned_names)
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for message in messages:
    length = len(message)
    print(f"Message: '{message}' | Length: {length}")
    if length > 10:
        print("Message is longer than 10 characters")
```

```python

logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0
for log in logs:
    if log == "ERROR":
        error_count += 1
print("Total ERROR entries:", error_count)
```
